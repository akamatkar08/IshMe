```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\akama\OneDrive\Desktop\Data Science Bootcamp\IshMe\Capstone 2 - Sales Forecasting\raw_data\vgsales_cleaned.csv")
vgsales = df.copy()
```

```python
vgsales.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
fig = sns.countplot(x=vgsales["Platform"], ax=ax, order=vgsales["Platform"].value_counts().index)

ax.set_xticklabels(fig.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
vgsales["Genre"].value_counts().plot.pie(labeldistance=1.1, autopct="%.2f%%", radius=1.5)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.countplot(x=vgsales["Year"], data=vgsales, ax=ax)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
vgsales = vgsales[vgsales["Year"] < 2017]
```

```python
vgsales.groupby("Year")["Rank"].median().sort_values(ascending=True).plot.bar(figsize=(10, 5))
```

```python
vgsales.groupby("Year")["Rank"].median().plot.bar(figsize=(10, 5))
```

```python
print("Number of unique publishers:", vgsales["Publisher"].unique().size)

top10_pub = (vgsales["Publisher"].value_counts().head(10)/vgsales.shape[0]*100).sum()
low500_pub = (vgsales["Publisher"].value_counts().tail(500)/vgsales.shape[0]*100).sum()

print("\nPercentage from total of top 10 publishers: {:.2f}%".format(top10_pub))

print("Percentage from total of intermediate 69 publishers: {:.2f}%".format(100-top10_pub-low500_pub))

print("Percentage from total of smallest 500 publishers: {:.2f}%".format(low500_pub))
      
print("\n\nTop 10 publishers :\n", 
    vgsales["Publisher"].value_counts().head(10))
```

```python
# On which platform the median VG was a hit
vgsales.groupby("Platform")["Global_Sales"].median().sort_values(ascending=False).plot.bar(figsize=(10, 5),
                                                                                     ylabel="Global Sales in millions of copies per VG (median)")
```

```python
# On which platform the total VG sales were the best
vgsales.groupby("Platform")["Global_Sales"].sum().sort_values(ascending=False).plot.bar(figsize=(10, 5),
                                                                                   ylabel="Global Sales in millions of copies")
```

```python
# What are the best Genre for Global_Sales in general
vgsales.groupby("Genre")["Global_Sales"].sum().sort_values(ascending=False).plot.bar(figsize=(10, 5),
                                                                                   ylabel="Global Sales in millions of copies")
plt.show()
# What are the best Genre for Global_Sales (median video games sales)
vgsales.groupby("Genre")["Global_Sales"].median().sort_values(ascending=False).plot.bar(figsize=(10, 5),
                                                                                   ylabel="Global Sales in millions of copies per VG (median)")
plt.show()
```

```python
# What are the Top 10 Publishers for Global_Sales in general
vgsales.groupby("Publisher")["Global_Sales"].sum().sort_values(ascending=False).head(10).plot.bar(figsize=(15, 5),
                                                                                   ylabel="Global Sales in millions of copies")
plt.show()
# What are the Top 10 Publishers  for Global_Sales (median video games sales)
vgsales.groupby("Publisher")["Global_Sales"].median().sort_values(ascending=False).head(10).plot.bar(figsize=(15, 5),
                                                                                   ylabel="Global Sales in millions of copies per VG (median)")
plt.show()
```

```python
# What games did these publishers released ? Except Valve (Half-Life) and Maxis (the Sims, Sim City)
# they are mainly unknown to the average gamer. So they should basically be "One Trick Poney" publishers
# ie : publishers with only one or two games that were blockbusters.
top10_pub_median = vgsales.groupby("Publisher")["Global_Sales"].median().sort_values(ascending=False).head(10)
```
