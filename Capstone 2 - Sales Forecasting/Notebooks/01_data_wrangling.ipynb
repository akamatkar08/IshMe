```python
# Numerical features DataFrame
df_num = df.select_dtypes(('int', 'float'))

df_num.head()
```

```python
# Get column names from categorical dataset
categorical_features = df_cat.columns
print(categorical_features)

print('-' * 60)

# Get column names from numerical dataset
numerical_features = df_num.columns
print(numerical_features)
```

```python
# Checking most repetitive values

for cat_name in categorical_features:
    
    print('-' * 50)
    print("Column Name: ", cat_name)
    print(' ' * 50)
    
    print(df[cat_name].value_counts().head())
    
    print('-' * 50)
```

```python
# Check number of missing values
df.isna().sum()
```

```python
# Filter and view feature descriptions
df[['Year', 'Publisher']].describe(include='all')
```

```python
# Fill missing values of year feature
df.Year = df.Year.fillna(df.Year.mean())
```

```python
# Change datatype to integer
df.Year = df.Year.astype('int32')

df.Year
```

```python
# Fill missing values in Publisher feature
df.Publisher.value_counts(normalize=True)
```

```python
# Replace with EA
df.Publisher = df.Publisher.fillna(df.Publisher.mode()[0])
```
