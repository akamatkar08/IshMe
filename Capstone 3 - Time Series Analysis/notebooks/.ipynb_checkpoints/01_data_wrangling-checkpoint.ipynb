## Importing Data<a id='Importing Data'></a>

Importing the datasets as variables for analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# from library.sb_utils import save_file

# Reading CSV data from COLevelsByAtmosphere, GloballandTemperaturesByCountry
CO_data = pd.read_csv(r"C:\Users\akama\OneDrive\Desktop\Data Science Bootcamp\IshMe\Capstone 3 - Time Series Analysis\raw_data\COLevelsByAtmosphere.csv")
country_data = pd.read_csv(r"C:\Users\akama\OneDrive\Desktop\Data Science Bootcamp\IshMe\Capstone 3 - Time Series Analysis\raw_data\GlobalLandTemperaturesByCountry.csv")
global_data = pd.read_csv(r"C:\Users\akama\OneDrive\Desktop\Data Science Bootcamp\IshMe\Capstone 3 - Time Series Analysis\raw_data\GlobalTemperatures.csv")
CO_emissions = pd.read_excel(r"C:\Users\akama\OneDrive\Desktop\Data Science Bootcamp\IshMe\Capstone 3 - Time Series Analysis\raw_data\CO2Data.xlsx")
```

## First Dataset<a id='First Dataset'></a>

First dataset is compiled from datasets acquired from Berkeley Earth. 

https://www.kaggle.com/derevirn/data-analysis-on-climate-change/data

https://www.kaggle.com/berkeleyearth/climate-change-earth-surface-temperature-data

```python
# Deriving data only pertaining to the US
US_data = country_data[country_data["Country"] == "United States"].copy()

# Convering "dt" column to datetime object
US_data["dt"] = pd.to_datetime(US_data["dt"])

# Filtering US_data 
US_data["Month"] = US_data["dt"].dt.month
US_data["Year"] = US_data["dt"].dt.year
US_data = US_data[(US_data["Year"] > 1957) & (US_data["Year"] < 2014)]
US_data = US_data.reset_index(drop=True)
US_data = US_data.rename(columns={"AverageTemperature" : "AverageTemperature_US", "AverageTemperatureUncertainty" : "AverageTemperatureUncertainty_US"})
US_data = US_data[["Year", "Month", "AverageTemperature_US", "AverageTemperatureUncertainty_US"]]
# US_data
```

```python
# Filtering global_data
global_data["dt"] = pd.to_datetime(global_data["dt"])
global_data["Month"] = global_data["dt"].dt.month
global_data["Year"] = global_data["dt"].dt.year
global_data = global_data[(global_data["Year"] > 1957) & (global_data["Year"] < 2014)]
global_data = global_data.reset_index(drop=True)
global_data = global_data[["Year", "Month", "LandAverageTemperature", "LandAverageTemperatureUncertainty", "LandMaxTemperature",
                          "LandMaxTemperatureUncertainty", "LandMinTemperature", "LandMinTemperatureUncertainty", "LandAndOceanAverageTemperature"]]

# global_data
```

```python
# Filtering CO Levels data
CO_data = CO_data[CO_data["Year"] < 2014]

# Final dataset for analysis
full_data = CO_data.merge(global_data, on=["Year", "Month"])
full_data = full_data.merge(US_data, on=["Year", "Month"])
full_data = full_data.drop(columns="Decimal Date")
full_data
```

The columns Carbon Dioxide (ppm), Seasonally Adjusted CO2 (ppm), Carbon Dioxide Fit (ppm), Seasonally Adjusted CO2 Fit (ppm) all have missing values. These can be replaced with mean values for each column

```python
print(full_data[~full_data["Carbon Dioxide (ppm)"].notnull()].index)
print(full_data[~full_data["Seasonally Adjusted CO2 (ppm)"].notnull()].index)
print(full_data[~full_data["Carbon Dioxide Fit (ppm)"].notnull()].index)
print(full_data[~full_data["Seasonally Adjusted CO2 Fit (ppm)"].notnull()].index)
```

```python
january = full_data[full_data["Month"] == 1]
february = full_data[full_data["Month"] == 2]
march = full_data[full_data["Month"] == 3]
april = full_data[full_data["Month"] == 4]
june = full_data[full_data["Month"] == 6]
october = full_data[full_data["Month"] == 10]

print("\nJanuary:", january["Carbon Dioxide (ppm)"].mean())
print("February:", february["Carbon Dioxide (ppm)"].mean())
print("March:", march["Carbon Dioxide (ppm)"].mean())
print("April:", april["Carbon Dioxide (ppm)"].mean())
print("June:", june["Carbon Dioxide (ppm)"].mean())
print("October:", october["Carbon Dioxide (ppm)"].mean())

full_data.loc[0, "Carbon Dioxide (ppm)"] = january["Carbon Dioxide (ppm)"].mean()
full_data.loc[1, "Carbon Dioxide (ppm)"] = february["Carbon Dioxide (ppm)"].mean()
full_data.loc[5, "Carbon Dioxide (ppm)"] = june["Carbon Dioxide (ppm)"].mean()
full_data.loc[9, "Carbon Dioxide (ppm)"] = october["Carbon Dioxide (ppm)"].mean()
full_data.loc[73, "Carbon Dioxide (ppm)"] = february["Carbon Dioxide (ppm)"].mean()
full_data.loc[74, "Carbon Dioxide (ppm)"] = march["Carbon Dioxide (ppm)"].mean()
full_data.loc[75, "Carbon Dioxide (ppm)"] = april["Carbon Dioxide (ppm)"].mean()
```

```python
january = full_data[full_data["Month"] == 1]
february = full_data[full_data["Month"] == 2]
march = full_data[full_data["Month"] == 3]
april = full_data[full_data["Month"] == 4]
june = full_data[full_data["Month"] == 6]
october = full_data[full_data["Month"] == 10]

print("January:", january["Seasonally Adjusted CO2 (ppm)"].mean())
print("February:", february["Seasonally Adjusted CO2 (ppm)"].mean())
print("March:", march["Seasonally Adjusted CO2 (ppm)"].mean())
print("April:", april["Seasonally Adjusted CO2 (ppm)"].mean())
print("June:", june["Seasonally Adjusted CO2 (ppm)"].mean())
print("October:", october["Seasonally Adjusted CO2 (ppm)"].mean())

full_data.loc[0, "Seasonally Adjusted CO2 (ppm)"] = january["Seasonally Adjusted CO2 (ppm)"].mean()
full_data.loc[1, "Seasonally Adjusted CO2 (ppm)"] = february["Seasonally Adjusted CO2 (ppm)"].mean()
full_data.loc[5, "Seasonally Adjusted CO2 (ppm)"] = june["Seasonally Adjusted CO2 (ppm)"].mean()
full_data.loc[9, "Seasonally Adjusted CO2 (ppm)"] = october["Seasonally Adjusted CO2 (ppm)"].mean()
full_data.loc[73, "Seasonally Adjusted CO2 (ppm)"] = february["Seasonally Adjusted CO2 (ppm)"].mean()
full_data.loc[74, "Seasonally Adjusted CO2 (ppm)"] = march["Seasonally Adjusted CO2 (ppm)"].mean()
full_data.loc[75, "Seasonally Adjusted CO2 (ppm)"] = april["Seasonally Adjusted CO2 (ppm)"].mean()

full_data.iloc[[0, 1, 5, 9, 73, 74, 75], :]
```

```python
full_data[~full_data["Carbon Dioxide Fit (ppm)"].notnull()]

january = full_data[full_data["Month"] == 1]
february = full_data[full_data["Month"] == 2]

print("January:", january["Carbon Dioxide Fit (ppm)"].mean())
print("February:", february["Carbon Dioxide Fit (ppm)"].mean())

full_data.loc[0, "Carbon Dioxide Fit (ppm)"] = january["Carbon Dioxide Fit (ppm)"].mean()
full_data.loc[1, "Carbon Dioxide Fit (ppm)"] = february["Carbon Dioxide Fit (ppm)"].mean()
```

```python
full_data[~full_data["Seasonally Adjusted CO2 Fit (ppm)"].notnull()]

january = full_data[full_data["Month"] == 1]
february = full_data[full_data["Month"] == 2]

print("January:", january["Seasonally Adjusted CO2 Fit (ppm)"].mean())
print("February:", february["Seasonally Adjusted CO2 Fit (ppm)"].mean())

full_data.loc[0, "Seasonally Adjusted CO2 Fit (ppm)"] = january["Seasonally Adjusted CO2 Fit (ppm)"].mean()
full_data.loc[1, "Seasonally Adjusted CO2 Fit (ppm)"] = february["Seasonally Adjusted CO2 Fit (ppm)"].mean()

full_data.iloc[[0, 1], :]
```

```python
full_data['Date'] = pd.to_datetime(full_data[['Year', 'Month']].assign(Day=1))
date = full_data['Date']
full_data.drop(labels=['Date', 'Year', 'Month'], axis=1, inplace=True)
full_data.insert(0, 'Date', date)
```

## Second Dataset<a id='Second Dataset'></a>

This is the second dataset. 

https://github.com/owid/co2-data/blob/master/owid-co2-codebook.csv

```python
# Filtering for 
CO_emissions = CO_emissions[(CO_emissions["country"] == "United States") & ((CO_emissions["year"] > 1957) & (CO_emissions["year"] < 2014))]
CO_emissions = CO_emissions.reset_index()
CO_emissions.drop(columns="index", inplace=True)
```

```python
CO_emissions.head()
```

```python
# Checking percentage of NaN values per column
CO_emissions.isna().sum() * 100 / len(CO_emissions)

# Copying columns with less than 50% NaN values
add_columns = CO_emissions[["primary_energy_consumption", "energy_per_capita", "energy_per_gdp"]]

# Dropping and rejoining DFs
CO_emissionsDropNA = CO_emissions.dropna(axis=1)
CO_emissionsDropNA = CO_emissionsDropNA.join(add_columns)
CO_emissions = CO_emissionsDropNA
```

```python
print(CO_emissions.isna().sum())
```

```python
CO_emissions.drop(columns=['iso_code', 'cumulative_cement_co2', 'cumulative_coal_co2', 'cumulative_flaring_co2',
                          'cumulative_gas_co2', 'cumulative_oil_co2', 'share_global_cumulative_cement_co2',
                          'share_global_cumulative_coal_co2', 'share_global_cumulative_flaring_co2',
                          'share_global_cumulative_gas_co2', 'share_global_cumulative_oil_co2'], inplace=True)
```
