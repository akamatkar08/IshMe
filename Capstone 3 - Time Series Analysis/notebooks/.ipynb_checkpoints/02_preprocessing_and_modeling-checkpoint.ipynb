# Capstone 3 - Time Series Analysis

The general overview of the following notebook is to analyze each dataset separately. The temperature data will be forecasted using a seasonal ARIMA model from the statsmodels package and the carbon emission dataset will need a multivariate LSTM Neural Network using Tensorflow and Keras.

**Dataset 1**

The temperature data will be decomposed to check for trend, residuals, and seasonality. If they exist, differencing will be used to remove these changes from the time series so all that's left is white noise. By the Dickey-Fuller test, if the data is stationary, then an SARIMAX model can be implemented to forecast the data. 

**Dataset 2**

The carbon emission data will require feature engineering to remove unnecessary columns. A function borrowed from machinelearningmastery.com will be used to convert the time series into a supervised learning problem, where the current data points at time t will be used to forecast the per capita carbon emission in the next year.

#### Decomposition

Decomposing the graph shows that the dataset has fairly strong seasonality, but it doesn't have much effect on each value.

```python
# Decomposing graph
decomp = seasonal_decompose(world_data['LandOcean_MovAvg'], model='additive', period=12)

fig, ax = plt.subplots(3,1, figsize=(10, 7))

# Plotting decomposition
ax[0].plot(decomp.seasonal)
ax[1].plot(decomp.trend)
ax[2].plot(decomp.resid)
ax[0].set_title('Seasonality')
ax[1].set_title('Trend')
ax[2].set_title('Residual')

plt.tight_layout()
plt.show()
```

#### Differencing
Both seasonal and non-seasonal differencing are applied and, the plot below looks like white noise with the p-value being much lower than 0.05.

```python
# Non-seasonal and seasonal differencing
both_diff = world_data.diff().diff(12).dropna()

# Plotting differencing
fig, ax = plt.subplots(figsize=(15, 3))
ax.plot(both_diff.LandOcean_MovAvg)
ax.set_title('Both Differencing')

plt.show()

# Dickey-Fuller test
result = adfuller(both_diff['LandOcean_MovAvg'])
adtest = pd.Series(result[0:4], index=['Test Statistic', 'p-value', 'Lags Used', 'Observations'])

print('Stationarity Test Seasonal Differencing:', adtest)
```

#### Autocorrelation
The tapering off in the graphs of the non-seasonal plots below show that both the AR and MA parts will be present in the non-seasonal component. In the seasonal plots, only the MA plot will be there since there is a significant drop off at q=2.

```python
fig, ax = plt.subplots(2,2, figsize=(13, 5))

# Plotting acf and pacf to find other values for non-seasonal
plot_acf(both_diff['LandOcean_MovAvg'], lags=11, zero=False, alpha=0.05, title='Non-Seasonal Autocorrelation', ax=ax[0,0])
plot_pacf(both_diff['LandOcean_MovAvg'], lags=11, zero=False, alpha=0.05, title='Non-Seasonal Partial Autocorrelation', ax=ax[0,1])

# Seasonal autocorrelation
plot_acf(both_diff['LandOcean_MovAvg'], lags=[12, 24, 36, 48, 60], zero=False, alpha=0.05, title='Seasonal Autocorrelation', ax=ax[1,0])
plot_pacf(both_diff['LandOcean_MovAvg'], lags=[12, 24, 36, 48, 60], zero=False, alpha=0.05, title='Seasonal Partial Autocorrelation', ax=ax[1,1])

plt.tight_layout()
plt.show()
```

## Modeling
After testing several combinations of p and q values between 1 and 3, the model with the lowerst AIC and the best values for P > |z| was (1, 1, 1)(0, 1, 2, 12). The summary of the in-sample predictions are then evaluated.

```python
# p, d, q, M, seasonal order components
order = (1, 1, 1) 
sorder = (0, 1, 2, 12)

# Building model
sarimax = SARIMAX(world_data['LandOcean_MovAvg'].loc['1975-04':], order=order, seasonal_order=sorder)

t_model = sarimax.fit(maxiter=200)

t_model.summary()
```

#### In-sample Predictions
The mean absolute percent error is calculated for the in-sample predictions to view how accurate the forecast is for data that we already have. Since the returned value is very low at 0.04%, this model appears to be accurate enough to be used to forecast future values.

```python
# Forecast predictions
forecast = t_model.get_prediction(start=-240)

# Plot predictions
fig1 = px.line(world_data, x=world_data.index, y=world_data.LandOcean_MovAvg, width=750, height=400, 
               labels={'LandOcean_MovAvg':'Temp. (C)'}, title='In-Sample Predictions')
fig1.add_scatter(x=forecast.predicted_mean.index, y=forecast.predicted_mean, mode='lines', name='Predicted')
fig1.add_scatter(x=forecast.predicted_mean.index, y=forecast.conf_int().iloc[:,0], name='Lower Bound')
fig1.add_scatter(x=forecast.predicted_mean.index, y=forecast.conf_int().iloc[:,1], fill='tonexty', name='Upper Bound')
fig1.show()

# Mean absolute percent error
def mape(fc, true):
    mape = np.mean((np.abs(fc-true)/np.abs(true)))
    return mape

acc = mape(forecast.predicted_mean, world_data['LandOcean_MovAvg'].iloc[-240:])*100
print('Mean Absolute Percent Error:', acc)
```

```python
t_model.plot_diagnostics(figsize=(15, 10))
plt.show()
```

#### Transforming the Data
The following function has been taken from machinelearningmastery.com and serves the purpose of transforming a time series analysis problem into a supervised learning algorithm through the use of the shift() function.

```python
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg
```

The data values are converted to a float datatype and the MinMaxScaler function is used to normalize the data between the values of 0 and 1. Since there are no categorical features, no label encoding is necessary. The previous function is also called with the arguments 0 to represent the time step chosen for the inputs and 2 for the output.

```python
# Converting dataset values to float32
dataset = df2
values = df2.values
values = values.astype('float32')

# Scaling dataset values from 0 to 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)
reframed = series_to_supervised(scaled, 1, 2)

# Number of lag years
n_years = 1
n_features = 24
```

All the columns except for the input and ouput data need to be dropped. The dataset below is shown to have 13 features in total, with the first 12 being the input and the last one being the output. This means the current data (t) will be used to predict next years carbon emissions (t+1)

## Modeling
The dataset can now be split into testing and training sets for the purpose of modeling. For this particular project, the most recent 30 years of data will be used to train the dataset, with the remaining data as the test set. This is because the emissions in modern society are more applicable than emissions from the early 1900s.

The inputs and outputs are then assigned by their time step value and reshaped into the proper 3D format necessary for an LSTM model. 

```python
# Split into train and test sets
values = reframed.values
n_train_years = 30
train = values[:-n_train_years, :]
test = values[-n_train_years:, :]

# Split into input and outputs
n_obs = n_years * n_features
train_X, train_y = train[:, :n_obs], train[:, -n_features]
test_X, test_y = test[:, :n_obs], test[:, -n_features]

# Reshape input to be 3D [samples, timesteps, features]
train_X = train_X.reshape((train_X.shape[0], n_years, n_features))
test_X = test_X.reshape((test_X.shape[0], n_years, n_features))
print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)
```

#### Defining the Model
The LSTM model can then be defined with 75 neurons in the first layer and the input shape defined from the training sets. 

```python
# design network
model = Sequential()
model.add(LSTM(75, input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')

# early stoppings
model_path = 'Bidirectional_LSTM_Multivariate.h5'
early_stopings = EarlyStopping(monitor='val_loss', min_delta=0, patience=0, verbose=1, mode='min')
checkpoint =  ModelCheckpoint(model_path, monitor='val_loss', save_best_only=True, mode='min', verbose=0)
callbacks = [early_stopings,checkpoint]

# fit network
history = model.fit(train_X, train_y, epochs=50, batch_size=72, validation_data=(test_X, test_y), callbacks=callbacks, verbose=2, shuffle=False)

# plot history
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
# make a prediction
testPredict = model.predict(test_X)
test_X = test_X.reshape((test_X.shape[0], n_years*n_features))

# invert scaling for forecast
inv_Forecast = np.concatenate((testPredict, test_X[:, -11:]), axis=1)
inv_Forecast = scaler.inverse_transform(inv_Forecast)
inv_Forecast = inv_Forecast[:,0]

# invert scaling for actual
test_y = test_y.reshape((len(test_y), 1))
inv_Actual = np.concatenate((test_y, test_X[:, -11:]), axis=1)
inv_Actual = scaler.inverse_transform(inv_Actual)
inv_Actual = inv_Actual[:,0]

# calculate RMSE
mae = mean_absolute_error(inv_Actual, inv_Forecast)
rmse = np.sqrt(mean_squared_error(inv_Actual, inv_Forecast))
print('Test RMSE: %.3f' % rmse)
print('Test MAE: %.3f' % mae)
```

```python
print(inv_Actual)
print(inv_Forecast)
```

```python
# Calculating R^2 value
corr_matrix = np.corrcoef(inv_Actual, inv_Forecast)
corr = corr_matrix[0,1]
R_sq = corr**2

# X-axis values
aa = df2.iloc[-30:].index

# Plotting actual vs. prediction
plt.figure(figsize=(8,4))

plt.plot(aa, df2['co2_per_capita'].iloc[-30:], marker='.', label='Actual')
plt.plot(aa, inv_Forecast, 'r', label='Forecast')
plt.tight_layout()

plt.ylabel('CO2 emissions per capita', size=12)
plt.xlabel('Year', size=12)
plt.title('Actual vs. Forecasted Carbon Emissions', size=12)
plt.legend(fontsize=12)
plt.show();

print('R Squared Score:', R_sq)
print('Forecasted CO2 Emission Per Capita in 2020:', inv_Forecast[-1])
```
