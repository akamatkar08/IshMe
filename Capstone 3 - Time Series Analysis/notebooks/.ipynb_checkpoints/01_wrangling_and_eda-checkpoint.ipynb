## Dataset 1 - Wrangling and Cleaning
The first dataset has been compiled and repackaged from Berkeley Earth who have combined 1.6 billion temperature reports from 16 pre-existing archives. While the dataset is a bit old and stops at 2015, this can actually be used to our benefit to check the accuracy of the predictive model against real world data from recent years.

```python
# Change 'dt' column to datetime
world_data = raw_data.copy()
world_data["dt"] = pd.to_datetime(world_data["dt"])
world_data["Date"] = pd.to_datetime(world_data.dt, format="%Y")

# Filter dataset to 1850 onward
world_data = world_data[(world_data["Date"].dt.year > 1850)]
world_data = world_data.reset_index(drop=True)
world_data = world_data[["Date", "LandAverageTemperature", "LandAndOceanAverageTemperature"]]

# Check dataset
world_data
```

```python
# Drop NA values
world_data.dropna(inplace=True)

# Calculate rolling mean by year and add as new columns
world_data["Land_MovAvg"] = world_data["LandAverageTemperature"].rolling(12).mean()
world_data["LandOcean_MovAvg"] = world_data["LandAndOceanAverageTemperature"].rolling(12).mean()

world_data.dropna(inplace=True)

# Subsetting column into separate DF
world_data_ts = world_data[["LandAverageTemperature"]]
```

```python
# Saving dataset before setting index
world_data.to_csv(r"C:\Users\akama\OneDrive\Desktop\Data Science Bootcamp\IshMe\Capstone 3 - Time Series Analysis\data\processed\world_data.csv", index= False)
```

```python
# Setting index as 'Date' column
world_data.set_index("Date", inplace=True)
```

```python
# Checking index
world_data.head()
```

## EDA

The general consensus within the scientific community is that a 1.5 degree celsius from pre-industrial revolution levels would have disastrous consequences around the world and a 2 degree increase would essentially be the point of no return. The historical data will be used to see how close we are (or were in 2015) to those values.

The following graph plots the 12 month rolling average for each month from 1850 to 2015. The graph shows an upward trend, signifying an increase with a somewhat weak seasonal trend. 

```python
# Plotting 12 month rolling average for each month
fig1 = px.line(world_data, x=world_data.index, y=world_data.LandOcean_MovAvg, 
               title="12 Month Global Land & Ocean Temp. Rolling Avg", labels={"LOMA_Temp": "Temp. (Celsius)"})

# Pre industrial revolution mean
pre_mn = world_data["LandOcean_MovAvg"].loc[:"1900-01-01"].mean()

fig1.add_scatter(x=["1850-12-01", "2015-12-01"], y=[pre_mn+1.5, pre_mn+1.5], mode="lines", name="1.5 Degree Increase")
fig1.add_scatter(x=["1850-12-01", "2015-12-01"], y=[pre_mn+2, pre_mn+2], mode="lines", name="2 Degree Increase")
fig1.show()

print('Pre-Industrial Mean Temp:', round(pre_mn, 3), 'Degrees Celsius')
print('Danger Benchmark 1:', round(pre_mn+1.5, 3), 'Degrees Celsius')
print('Danger Benchmark 2:', round(pre_mn+2, 3), 'Degrees Celsius')
```

## Dataset 2 - Wrangling and Cleaning

This second dataset is a collection of key metrics maintained by 'Our World in Data' and includes data on CO2 emissions (annual, per capita, cumulative and consumption-based), other greenhouse gases, energy mix, and other relevant metrics. This dataset will be filtered to only evaluate the United State's data.

```python
# Filtering dataset for United States only, from 1900 onwards
CO_emissions = CO_emissions[(CO_emissions["country"] == "United States") & ((CO_emissions["year"] > 1899))]
CO_emissions = CO_emissions.reset_index()
CO_emissions.drop(columns="index", inplace=True)
```

```python
CO_emissions.head()
```

There are numerous metrics with too many missing values to be usable for analysis. The features with more than half the values missing would be better off being removed as they don't contain enough information to be beneficial.

```python
# Checking percent of missing values per column and creating DF
percent_missing = CO_emissions.isna().sum() * 100 / len(CO_emissions)
missing_df = pd.DataFrame({"percent_missing": percent_missing})

# Showing columns with missing values only
missing_df[missing_df["percent_missing"] != 0]
```

Some of the features actually have less than 50% of their values missing, such as the carbon emission data from the flaring industry, and will be retained to see if the information is valuable.

```python
# Copying columns with less than 50% NaN values 
add_columns = CO_emissions[["co2_per_gdp", "co2_per_unit_energy", "flaring_co2", 
                            "flaring_co2_per_capita", "share_global_flaring_co2", "gdp"]]

# Dropping and rejoining DFs
CO_emissionsDropNA = CO_emissions.dropna(axis=1)
CO_emissionsDropNA = CO_emissionsDropNA.join(add_columns)
CO_emissions = CO_emissionsDropNA
```

```python
CO_emissions.head()
```

```python
# Checking percent of missing values per column and creating DF
percent_missing = CO_emissions.isna().sum() * 100 / len(CO_emissions)
missing_df = pd.DataFrame({"percent_missing": percent_missing})

# Showing columns with missing values only
missing_df[missing_df["percent_missing"] != 0]
```

Since there are still over 37 columns in the DataFrame, the columns deemed unnecessary must be dropped manually through intuition. Namely, the cumulative data appears to be redundant both in referance to the global percentage and in general.

```python
CO_emissions.columns
```

```python
# Dropping unnecessary columns
columns = ["iso_code", "cumulative_co2", "cumulative_cement_co2", "cumulative_coal_co2", "cumulative_gas_co2", 
           "cumulative_oil_co2", "share_global_cumulative_cement_co2", 
           "share_global_cumulative_coal_co2", "share_global_cumulative_gas_co2", "share_global_cumulative_oil_co2"]
```

```python
CO_emissions.drop(columns=columns, inplace=True)
```

```python
# Subsetting for columns with most missing values
NaN_values_DF = CO_emissions[["year", "co2_per_unit_energy", "flaring_co2", "flaring_co2_per_capita", "share_global_flaring_co2"]]

# Checking DF for rows with NaN values
NaN_values_DF[NaN_values_DF.isna().any(axis=1)]
```

```python
# Dropping co2_per_unit_energy and share_global_flaring_co2
NaN_values_DF.drop(columns=["co2_per_unit_energy", "share_global_flaring_co2"], inplace=True)
```

```python
# Filtering for rows where entries are not null
NaN_values_DF[~NaN_values_DF.isna().any(axis=1)]
```

Since the per capita emissions of the flaring industry may potentially be important enough to retain yet has a substantial percentage of its values missing, the data from the past 50 years will be inverted and used to replace the missing values. This is due to the fact that there seems to be a trend of it steadily decreasing and increasing, which would make sense because the value is relative to the population.

```python
# Create series from flaring_co2_per_capita values for 50 years
ts = pd.Series(NaN_values_DF["flaring_co2_per_capita"].iloc[50:101].values)

# Reverse the values of column, keeping index the same
ts = pd.Series(ts.values[::-1], ts.index)
ts.head()
```

```python
# Setting the first 50 rows of DF to above series
CO_emissions["flaring_co2_per_capita"].iloc[:51] = ts
```

```python
# Dropping columns deemed unnecessary from full dataset
CO_emissions.drop(columns=["co2_per_unit_energy", "flaring_co2", "share_global_flaring_co2"], inplace=True)
```

Finally, the dataset to check to see if any of the rows still have missing values.

```python
# Checking if last 2 columns still have null values
CO_emissions[CO_emissions["co2_per_gdp"].isna()]
```

```python
# Forward filling last missing value
CO_emissions.fillna(method="ffill", inplace=True)
```

Since we know that the dataset pertains to the United States only and this value is categorical, it can be dropped to keep all the datatypes in the DataFrame consistent. 

```python
# Dropping country name column
CO_emissions.drop(columns='country', inplace=True)
```

```python
CO_emissions.columns
```

```python
# Reordering columns of DF
column_order = ["year", "co2", "co2_growth_abs", "share_global_co2", 
                "coal_co2", "cement_co2", "gas_co2", "oil_co2", 
                "co2_per_capita", "coal_co2_per_capita", "cement_co2_per_capita", 
                "gas_co2_per_capita", "oil_co2_per_capita", "flaring_co2_per_capita", 
                "share_global_coal_co2", "share_global_cement_co2", "share_global_gas_co2",
                "share_global_oil_co2", "population", "gdp"]

CO_emissions = CO_emissions[column_order]

# Setting year column to datetime value and as index
CO_emissions['year'] = pd.to_datetime(CO_emissions.year, format='%Y')
df2 = CO_emissions.copy()
df2.set_index('year', inplace=True)
```

## EDA
The industrial carbon emission data will be analyzed to see which particular industries have the highest values and therefore the most impact on the overall carbon emission in the country. Each feature of the dataset will also be plotted to see if there is any discernable trend within the data.

```python
# Plotting CO2 emissions by industry in million tons and per capita
df_emissions = pd.DataFrame(df2, columns=["co2", "coal_co2", "cement_co2", "gas_co2", "oil_co2"])
df_per_capita = pd.DataFrame(df2, columns=["co2_per_capita", "coal_co2_per_capita", "cement_co2_per_capita", 
                                                    "gas_co2_per_capita", "oil_co2_per_capita", "flaring_co2_per_capita"])

# PLotting both DFs
df_emissions.plot(figsize=(10, 5))
plt.xlabel("Time (Yearly)")
plt.ylabel("Carbon Emissions By Industry (Million Tons)")

df_per_capita.plot(figsize=(10, 5))
plt.xlabel("Time (Yearly)")
plt.ylabel("Carbon Emissions Per Capita (Tons Per Person)")
```

One trend that is clear from the plots below is that the share CO2 emissions have been steadily decreasing. This can mean that either the US has been making efforts to reduce their impact, or as other countries develop, their impact increases so the percentage of total emissions from the US has been decreasing. Since the plots above show that the industrial impact per capita has been increasing over time, it's more likely that it is the latter of the two.

```python
df_plot = df2
df_plot

for i, col in enumerate(df_plot.columns):
    df_plot[col].plot(fig=plt.figure(i))
    plt.title(col)
    
plt.show()
```

Finally, a bar graph is plotted to view which industries have the largest impact on total CO2 emissions per capita.

```python
# Dropping total co2 emissions per capita and plotting industries on bar chart
co2_by_industry = df_per_capita.drop(columns=["co2_per_capita"]).sum().sort_values()
fig = plt.figure()
x_label = df_per_capita.columns
co2_by_industry.plot(kind="bar")
```
