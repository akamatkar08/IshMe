#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

```python
# Number of rows and columns
properties.shape

# First few rows of DF
properties.head()
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# Transpose DF and check head/index
tran_prop = properties.T
tran_prop.head()
tran_prop.index

# Reset index and reassign to transposed DF
tran_prop = tran_prop.reset_index()
tran_prop.index
tran_prop.head()

# Look at columns indexes and first row
tran_prop.columns
tran_prop.iloc[[0]]

# Set column indexes to values in first row
tran_prop.columns = tran_prop.iloc[0]
tran_prop.head()

# Drop row at index 0
tran_prop = tran_prop.drop(0)
tran_prop.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename columns 'Unnamed' and 'NaN' to 'London_Borough' and 'ID'
tran_prop = tran_prop.rename(columns = {'Unnamed: 0' : 'London_Borough', pd.NaT: 'ID'})
tran_prop.head()
tran_prop.columns
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt DF to make column indexes a single column
melted_prop = pd.melt(tran_prop, id_vars=['London_Borough', 'ID'])
melted_prop.head()

# Rename new columns from '0' and 'value' to 'Month' and 'Average_Price'
melted_prop = melted_prop.rename(columns = {0: 'Month', 'value': 'Average_Price'})
melted_prop.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Check data types of melted DF
melted_prop.dtypes

# Change data type of "Average_Price" column to float
melted_prop['Average_Price'] = pd.to_numeric(melted_prop['Average_Price'])
melted_prop.dtypes

# Check missing values
melted_prop.count()
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Check 'London_Borough' column to see all of the unique entries
melted_prop['London_Borough'].unique()

melted_prop[melted_prop['London_Borough'] == 'Unnamed: 34'].head()
melted_prop[melted_prop['London_Borough'] == 'Unnamed: 37'].head()

# Check number of null entries in 'ID' column
melted_prop[melted_prop['ID'].isna()]

# Filter melted_prop for .notna() values and assign to new DF
NaNFreeDF1 = melted_prop[melted_prop['Average_Price'].notna()]
NaNFreeDF1.head()

# Check number of rows with complete information
NaNFreeDF1.count()

# Drop null values from melted_prop and assign to new DF
NaNFreeDF2 = melted_prop.dropna()
NaNFreeDF2.head()
NaNFreeDF2.count()

# Check to see if null columns have been dropped
NaNFreeDF2['London_Borough'].unique()

# Check dimensions of all three DFs
print(melted_prop.shape)
print(NaNFreeDF1.shape)
print(NaNFreeDF2.shape, '\n')

# Create a list of strings that are not London boroughs
nonBoroughs = ['Inner London', 'Outer London', 'NORTH EAST', 'NORTH WEST', 
               'YORKS & THE HUMBER', 'EAST MIDLANDS', 'WEST MIDLANDS', 
               'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 'SOUTH WEST', 
               'England']

# Filter row values that are in nonBoroughs list
NaNFreeDF2[NaNFreeDF2.London_Borough.isin(nonBoroughs)]

# Filter row values that are NOT in nonBoroughs list and assign to DF
NaNFreeDF2[~NaNFreeDF2.London_Borough.isin(nonBoroughs)]
NaNFreeDF2 = NaNFreeDF2[~NaNFreeDF2.London_Borough.isin(nonBoroughs)]
NaNFreeDF2['London_Borough'].unique()
NaNFreeDF2.shape
NaNFreeDF2.head()

# Assign final DF to variable 'df' after resetting index for analysis
df = NaNFreeDF2.reset_index(drop=True)

# Go through 'Month' column of DF to find the years and add a new column 'Year' to DF
df['Year'] = df['Month'].apply(lambda t: t.year)
df = df[['London_Borough', 'ID', 'Year', 'Month', 'Average_Price']]

# Omit scientific notation of 'Average_Price' column
pd.set_option('display.float_format', '{:.2f}'.format)

# Check data types, indexes and first few rows of DF
df.dtypes
df.index
print(df.head())

# Subset original DF to create new DFs for the average price for a specific year
# Reset indexes of new DFs
avg_price_2018 = df[df['Year'] == 2018].groupby(by=['London_Borough', 'Year']).mean().reset_index().drop(columns='Year')
avg_price_2021 = df[df['Year'] == 2021].groupby(by=['London_Borough', 'Year']).mean().reset_index().drop(columns='Year')

# Check indexes of new DF
avg_price_2018.index

avg_price_both = avg_price_2018.merge(avg_price_2021, on='London_Borough', suffixes=('_2018', '_2021'))

# Check DFs for averages prices in 2018, 2021, and both
avg_price_2018.sort_values(by='Average_Price').head()
avg_price_2021.sort_values(by='Average_Price').tail()
print('\n', avg_price_both)
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Filter df for the Borough 'Camden' and assign to variable 'camden_prices'
camden_prices = df[df['London_Borough'] == 'Camden']

# Plot change in prices over time for houses in the 'Camden' Borough
ax = camden_prices.plot(kind='line', x='Month', y='Average_Price')

# Set y axis label to 'Price'
ax.set_ylabel('Price')

# Plot average prices for 2018 and 2021
bx = avg_price_both[['London_Borough', 'Average_Price_2018', 'Average_Price_2021']]
bx.plot(kind='bar').set_xticklabels(avg_price_both['London_Borough'])
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Function to find differences in house prices between 1998 and 2018
def create_price_ratio(d):
    # Assign to variables the average price in a specific borough in 1998 and 2018
    y1998 = float(d['Average_Price'][d['Year']==1998])
    y2018 = float(d['Average_Price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio

# Call function with the borough 'Barking & Dagenham' to test
create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])

# Create empty dictionary to store ratios of each borough
final = {}

# Loop that iterates through each borough and 
for i in dfg['London_Borough'].unique():
    # Parameter to use as argument for create_price_ratio function
    borough = dfg[dfg['London_Borough']==i]
    # Call function and add to dictionary using the iterator for the key value of the borough name
    final[i] = create_price_ratio(borough)
    
print(final)

# Create new DF using the 'final' dictionary
df_ratios = pd.DataFrame(final)
df_ratios.head()

# Transpose DF to make borough names a single column and reset the index
# Assign transposed DF to original DF
tran_df_ratios = df_ratios.T
df_ratios = tran_df_ratios.reset_index()
df_ratios.head()

# Rename columns to 'London_Borough' and '2018'
df_ratios.rename(columns={'index':'London_Borough', 0:'2018'}, inplace=True)
df_ratios.head()

# Greatest change in prices is determined by highest ratios
# Get the top 15 boroughs by sorting in descending order and assigning values to 'top15'
top15 = df_ratios.sort_values(by='2018', ascending=False).head(15)
print(top15)

# Plot top 15 boroughs with the greatest change in prices with a bar chart
ax = top15[['London_Borough', '2018']].plot(kind='bar')
ax.set_xticklabels(top15.London_Borough)
```
